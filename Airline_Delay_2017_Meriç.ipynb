```python
import pandas as pd

pd.set_option('display.max_columns', None)

df = pd.read_csv("2017.csv.zip")

df["FL_DATE"] = pd.to_datetime(df["FL_DATE"])
df.head()
```

```python
df = df.drop(columns=['Unnamed: 27'])
df.head()
```

```python
delay_cols = ['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY','DEP_DELAY','TAXI_OUT','WHEELS_OFF','CRS_ELAPSED_TIME','ACTUAL_ELAPSED_TIME','AIR_TIME','DISTANCE']
df[delay_cols] = df[delay_cols].fillna(0)
df.head()
```

```python
import pandas as pd
import numpy as np

def format_time_safe(val):
    if pd.isna(val):
        return np.nan

    # Convert to string and remove colons if present
    val_str_raw = str(val)
    val_str_cleaned = val_str_raw.replace(':', '').strip()

    # Try to convert to int. Use float first to handle cases like '900.0'.
    # If it cannot be converted to a number, it's an invalid time.
    try:
        val_int = int(float(val_str_cleaned))
        val_str_padded = str(val_int).zfill(4)
    except ValueError:
        return np.nan # If it cannot be converted to int, it's an invalid time

    # Now parse hours and minutes from the padded string
    if len(val_str_padded) != 4: # Ensure it's a 4 digit time
        return np.nan

    hours = int(val_str_padded[:2])
    minutes = int(val_str_padded[2:])

    # Handle '2400' as '00:00'
    if hours == 24 and minutes == 0:
        return "00:00"

    # Validate hours and minutes (allowing 24:00 to be handled above)
    if not (0 <= hours <= 23) or not (0 <= minutes <= 59):
        return np.nan

    return f"{hours:02d}:{minutes:02d}"

time_cols = ['CRS_DEP_TIME', 'DEP_TIME','WHEELS_OFF']
for col in time_cols:
    df[col] = df[col].apply(format_time_safe)
```
